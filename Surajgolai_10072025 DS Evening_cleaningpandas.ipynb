```python
mean_age = data['age'].mean()
data['age'] = data['age'].fillna(mean_age)
data['age'] = data['age'].astype(int)
```

```python
avg = round(data['balance'].mean(), 2)      
maxx = round(data['balance'].max(), 2) 

miss_index = data[data['balance'].isna()].index
data.loc[miss_index[:500], 'balance'] = avg
data.loc[miss_index[500:1000], 'balance'] = maxx
data.loc[miss_index[1000:], 'balance'] = 0.00
```
